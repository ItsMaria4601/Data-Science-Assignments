```python
import pandas as pd
import numpy as np
df=pd.read_csv("weather_data.csv")
df
```

```python
check_nan_value_temp=df['temperature'].isnull().sum()  #checking for null values in temperature column in dataframe in boolean dtype and if we add sum() it will tell us abut the total null values in that column 
print('NaN value in temperature column are'+ " " + str(check_nan_value_temp))


check_nan_value_wind=df['windspeed'].isnull().sum()
print('NaN value in windspeed column are'+ " " + str(check_nan_value_wind))


check_nan_value_event=df['event'].isnull().sum()
print('NaN value in event column are'+ " " + str(check_nan_value_event))#parsed the int value to string
```

```python
#Applying data imputation technique (to fill missing data)


#1. fillna technique

new_df=df.fillna(0)  #new_df is used because we dont want to overwrite our original dataframe so that we can apply more techniques
new_df
```

```python
#2. fillna with dictionary to fill the coloumn with specific values

new_df=df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event':'no event'
})

new_df
```

```python
#3. fillna method forward fill is used to carry the values forward in the missing column

new_df=df.fillna(method="ffill") 
new_df
```

```python
new_df['temperature'].mean()
```

```python
new_df['temperature'].median()
```

```python
new_df.corr(method='pearson')  #highly positive with temperature and temperature with wind is 60% positive moderate
```

```python
#Defining windspeed bins 0-6, 6-8, 8-9, 9-12

bins_wind=[5,8,9,12]
```

```python
#names for group names

group_names=['low','medium','high']
```

```python
new_df['Wind Intensity']=pd.cut(new_df['windspeed'],bins_wind,labels=group_names)
new_df
```

```python
#Defining temperature bins 28-32, 32-34, 34-40

temp_bin=[27,32,34,40]

group_temp=['cold','moderate','hot']
```

```python
new_df['Temperature Intensity']=pd.cut(new_df['temperature'],temp_bin,labels=group_temp)
new_df
```
