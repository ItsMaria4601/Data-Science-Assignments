```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from matplotlib import pyplot as plt
```

```python
#directly loaded dataset from seaborn library

iris=sns.load_dataset('iris') 
iris
```

```python
#because we needed only two parameters X and Y for Y=mx+c (Linear Regression Model)

iris=iris[['sepal_length','sepal_width','petal_length','petal_width']]
iris
```

```python
#making a scatter plot to check for model fitness

#multiple variables in a scatter plot

sns.set_style("darkgrid")
X1=iris['petal_length']
y=iris['petal_width']
X2=iris['sepal_width']
X3=iris['sepal_length']

plt.scatter(X1,y)
plt.scatter(X2,y)
plt.scatter(X3,y)
plt.xlabel("petal length, sepal length, sepal width")
plt.ylabel("petal width")

plt.grid()

```

```python
#splitting into training and testing dataset

#Scikit-learn’s train_test_split expects data in the form of features and target. In scikit-learn, a features matrix is a two-dimensional grid of data where rows represent samples and columns represent features. A target is what you want to predict from the data.

features = ['petal_length','sepal_length','sepal_width']
X = iris.loc[:, features]
y = iris.loc[:, ['petal_width']]

X_train, X_test,y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=23)

#since the X_Train and y_train are 1D arrays hence we will convert them to 2D array

X_train=np.array(X_train).reshape(-1,1)

X_test=np.array(X_test).reshape(-1,1)  #-1 are the as many no of rows in matrix and there is only 1 column

X_test  #printing the X_test variable
```

```python
#save linear regression model into a variable

lr=LinearRegression()

#Train model using training data set

lr.fit(X_train,y_train)
```

```python
c= lr.intercept_
c
```

```python
m=lr.coef_
m 
```

```python
#Y=mx+c

#1st method

y_pred_train= m*X_train + c
y_pred_train.flatten()


```

```python
#now testing the model

y_pred_test1=lr.predict(X_test)
y_pred_test1

```

```python
#check for best fit line in testing (this step can be done for training data too!)

sns.set_style("darkgrid")
plt.title("Testing")
plt.scatter(X_test,y_test, c='hotpink')
plt.plot(X_test,y_pred_test1, c='red')
plt.xlabel("petal length")
plt.ylabel("petal width")
```

```python
#Mean Squared Error

print('Mean Squared Error: %.2f' % mean_squared_error(y_test, y_pred_test1))

#R squared Error

print('R squared Error is %.2f' % r2_score(y_test,y_pred_test1))
```
